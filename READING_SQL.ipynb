```python
from sqlalchemy import create_engine
engine=create_engine('sqlite:////content/chinook.db')
conn=engine.connect()
```

```python
new=pd.read_sql_table('employees',conn , index_col='EmployeeId',parse_dates=(['BirthDate','HireDate']))
new.head()
```

```python
new.to_sql('employees2', conn , if_exists='replace')
```
