```python
X =np.arange(-10,11)
```

```python
plt.figure(figsize=(12,6))
plt.title('My first plot')
plt.subplot(1, 2, 1) #row ,columns , panel selected
plt.plot(X , X**2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend('X^2' , 'Vertical line')
plt.xlabel('X')
plt.ylabel('X**2')

plt.subplot(1, 2, 2)
plt.plot (X , X**3)
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend('X^3' , 'Horizontal line')
plt.xlabel('X')
plt.ylabel('X**3')
```

```python
fig, axes = plt.subplots(1, 1, figsize=(12, 6))  # Create figure and axes with figsize
axes.plot(
    X, (X ** 2), color='red', linewidth=3,  # Use X instead of x
    marker='o', markersize=8, label='X^2')
axes.plot(
    X, (X ** 3),linewidth=5, color='purple',marker='x', label='X^3')
  # Use X instead of x
plt.legend()
plt.show()
```

Making lines with different patterns

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(X, X+ 0, linestyle='solid', label='solid line')
axes.plot(X, X+ 1, linestyle='dashed' , label = 'dashed line')
axes.plot(X, X + 2, linestyle='dashdot', label = 'dashdot line')
axes.plot(X, X + 3, linestyle='dotted' , label ='dotted line');
axes.legend()

axes.set_title("My Nice Plot")
```

```python
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(X, X+ 0, '-og', label="solid green")
axes.plot(X, X + 1, '--c', label="dashed cyan")
axes.plot(X, X + 2, '-.b', label="dashdot blue")
axes.plot(X, X+ 3, ':r', label="dotted red")
axes.legend()

axes.set_title("My Nice Plot")
```

### **SCATTER PLOT**

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='jet')
plt.colorbar()
plt.show()
```

```python
fig=plt.figure(figsize=(14, 6))
ax1 = plt.subplot(1,2,1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='jet')
plt.colorbar()


ax2= plt.subplot(1,2,2)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()
plt.show()


```

### **HISTOGRAM PLOTTING**

```python
values = np.random.randn(1000)
plt.subplots(1,1,figsize=(12, 6))

plt.hist(values, bins=100, alpha=0.8,
          histtype='bar', color='pink',
          edgecolor='red')
plt.xlim(xmin=-5, xmax=10)

plt.show()
```
